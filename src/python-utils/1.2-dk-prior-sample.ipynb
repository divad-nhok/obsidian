```python
import diagplots
import os
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
parent_dir = '/Users/davidkohn/dev/obsidian/output/mhrw_synth_01_01/'
samples_fname = 'output.npz'

keys_of_interest = [
    'gravReadings',
    'magReadings'
]
sample_row_idx = 0
file_extension = '.csv'
mu_e = 0
sigma_e = 2
seed = 99
```

```python
fpath = os.path.join(parent_dir, samples_fname)
samples = np.load(fpath)

for key in samples.keys():
    vals = samples[key]
    n = vals.shape
    print(key, n)

np.random.seed(seed)
for key in keys_of_interest:
    sensor_samples = samples[key]
    fwd_model_obs = sensor_samples[sample_row_idx, :]
    e = np.random.normal(mu_e, sigma_e, fwd_model_obs.size)
    new_obs = fwd_model_obs + e
    fname = key + file_extension
    fpath = os.path.join(parent_dir, fname)
    np.savetxt(fpath, new_obs)
```
