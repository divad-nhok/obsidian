```python
import diagplots
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd

%matplotlib inline
#%matplotlib notebook
%reload_ext autoreload
%autoreload 2
```

```python
curr_dir = '/Users/davidkohn/dev/obsidian/output'
```

```python
######################
# theta vs. log likelihood plot
fig_width = 20
fig_height = 20
xstr = 'Theta{}'
ystr = 'Log-likelihood'
title_str = 'How the likelihood changes as theta{} is varied from -100 to +100'
save_str = 'theta{}_vs_loglikelihood.png'
fontsize = 20
start_slice_idx = 100
param_range = 8

for idx in range(1, param_range):
    print(idx)
    parent_dir = '/Users/davidkohn/dev/obsidian/output/grid_search_0{}/'.format(idx)
    samples_fname = 'gridsearch.npz'

    fpath = os.path.join(parent_dir, samples_fname)
    samples = np.load(fpath)

    theta = samples['thetas'][:, idx - 1]
    ll = samples['likelihoods'].sum(axis = 1)
    
    fig = plt.figure(figsize = (fig_width, fig_height))
    plt.plot(theta, ll)
    
    plt.xlabel(
        xstr.format(idx),
        fontsize = fontsize
    )
    plt.ylabel(
        ystr,
        fontsize = fontsize
    )
    plt.title(
        title_str.format(idx),
        fontsize = fontsize
    )
    out_path = os.path.join(
        curr_dir, 
        save_str.format(idx)
    )
    plt.savefig(
        out_path
    )
    plt.clf()
```
