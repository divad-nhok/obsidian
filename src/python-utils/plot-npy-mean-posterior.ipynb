```python
#exp_date = '08_16_2018'
exp_date = 'gascoyne_v4_run06'
#num_list = ['{:02}'.format(num) for num in range(2, 3)]
num_list = ['01']
dir_parent = '/Volumes/david_hd/obsidian/output/experiments/'
#dir_parent = '/Users/davidkohn/Desktop'
#dir_expnum = os.path.join(dir_parent, exp_date, exp_num)
dir_output = '/Users/davidkohn/Desktop/'
search_str = 'npy'
```

```python
def get_fname_list(dir_parent, search_str):
    fname_list = [
        os.path.join(dir_parent, fname)
        for fname in os.listdir(dir_parent) 
        if os.path.isfile(os.path.join(dir_parent, fname))
        and search_str in fname
    ]
    return(fname_list)
```

```python
# display layer posterior probability voxel plot
# to save use vv.screenshot: 
# (1) first run this cell then close the image; 
# (2) then run this cell again with app.Run() commented out
# (3) then run the cell below where the screenshot is saved
def run_geovis_app(
    layer_voxels,
    fname_save = None
):
    app = vv.use()
    vv.figure(1)
    vv.xlabel('Eastings (units)')
    vv.ylabel('Northings (units)')
    vv.zlabel('Depth (units)')
    a = vv.gca()
    a.camera.fov = 200
    a.daspect = 1, 1, -1
    t = vv.volshow(layer_voxels, cm=vv.CM_JET)
    vv.ColormapEditor(a)
    if fname_save:
        vv.screenshot(
            fname_save,
            sf = 2,
            bg = 'w'
        )
        vv.clf()
```

```python
def plot_layer_posteriors(
    fname_list,
    dir_output = '',
    img_save_format = 'png',
    fname_save_template_str = 'mean-posterior-layer-{:02}.{}'
):
    for layer_idx, fname in enumerate(fname_list):
        fname_img = fname_save_template_str.format(layer_idx, img_save_format)
        fname_save = os.path.join(dir_output, fname_img)
        print(fname)
        layer_voxels = np.load(fname)
        run_geovis_app(layer_voxels, fname_save)
```

```python
voxel_sub_dir = True
for exp_num in num_list:
    dir_expnum = os.path.join(dir_parent, exp_date, exp_num)
    if voxel_sub_dir:
        dir_expnum = os.path.join(dir_expnum, 'voxels')
    print(dir_expnum)
    fname_list = get_fname_list(dir_expnum, search_str)
    plot_layer_posteriors(fname_list, dir_expnum)
    print()
```
